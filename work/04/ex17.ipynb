```
from urllib.request import urlopen
from urllib import parse
```

```
#문화재 목록 
url = 'http://www.cha.go.kr/cha/SearchKindOpenapiList.do'
data = urlopen(url).read().decode('utf-8')

print(type(data))
```

```
import re

pt = r'<totalCnt>([\w]+)</totalCnt>'
dt = re.search(pt, data)

print(data[:1024])
print(dt.group(1))
```

```
#전처리
data1 = data.replace('<![CDATA[','')
data1 = data1.replace(']]>','')

pt = r'<ccbaMnm1>([\w ]+)</ccbaMnm1>'
#print(data1)
dt = re.findall(pt, data1)

for item in dt:
    print(item)
```

```
from bs4 import BeautifulSoup as bs 
```

```
bt = bs(data, 'lxml-xml')
```

```
print(bt.find('totalCnt').text)
```

```
print(bt.find('totalCnt').getText())
```

```
def show(ccbaKdcd, ccbaAsno, ccbaCtcd) :
    url = 'http://www.cha.go.kr/cha/SearchKindOpenapiDt.do?'
    url = url + 'ccbaKdcd=' + str(ccbaKdcd )
    url = url + '&ccbaAsno=' + str(ccbaAsno)
    url = url + '&ccbaCtcd=' + str(ccbaCtcd)   
    memoData = urlopen(url)

    mdata = bs(memoData, 'lxml-xml')
    content = mdata.find('content').text
    
    return content
```

```
items = bt.find_all('item')

#ccbaKdcd=11&ccbaAsno=00010000&ccbaCtcd=11

for item in items :
    ccbaMnm1=item.find('ccbaMnm1').text
    ccbaKdcd=item.find('ccbaKdcd').text
    ccbaAsno=item.find('ccbaAsno').text
    ccbaCtcd=item.find('ccbaCtcd').text
    print(f'[{ccbaMnm1}] \n{show(ccbaKdcd,ccbaAsno,ccbaCtcd)}')
    print('-'*50)
```
