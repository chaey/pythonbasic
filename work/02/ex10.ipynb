```python
#2019년 부산의 1월에서 12월 평균, 최저, 최고 기온
avg = [4.5,6.6,10.5,13.4,19.3,21.1,24.3,27,23.1,18.6,13,7]
min = [-4.4,-4.2,0.5,2.1,11,15.5,19.8,20.3,15.8,10.4,2,-2.8]
max = [13.7,16.5,21,23.4,30.1,29.7,30.9,35,30.5,28.8,24,16.6]
```

```python
x = list(range(len(avg)))
xm = [str(i+1)+'월' for i in x] 

#음수 기호 표시
plt.rcParams['axes.unicode_minus'] = False

plt.plot(xm, max, 'go-', label='최대')
plt.bar(xm, avg, label='평균')
plt.plot(xm, min, 'rx--', label='최소')

plt.title('3월 기온변화')
plt.legend()
plt.show()
```

```python
area = list(data.keys())
area
```

```python
value = list(data.values())
value
```

```python
pm10 = []
o3 = []
co = []

for item in value :
    pm10.append(item[0])
    o3.append(item[1])
    co.append(item[2])
    
print(pm10)
print(o3)
print(co)
```

```python
x = list(range(len(pm10)))

plt.figure(figsize=(10, 9))

plt.subplot(3,1,1)
plt.plot(pm10, color='g', marker ='o', linestyle='--', label='pm10')
plt.legend()

plt.subplot(3,1,2)
plt.plot(o3, color='r', marker ='x', linestyle='-', label='o3')
plt.xticks(x)
plt.legend()

plt.subplot(3,1,3)
plt.plot(co, color='b', marker ='^', linestyle='-.', label='co')
plt.xticks(x, area)
plt.legend()

plt.show()
```

# 해결문제

```python
with open('지역평균기온.txt', 'r', encoding='utf-8') as fp:
    data = fp.readlines() 

area = []
temp = []

for line in data:
    line = line.replace('\n', '')
    item = line.split(',')
    
    area.append(item[0])
    temp.append(int(item[1]))
    
print(area)
print(temp)
```

```python
plt.bar(area, temp)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(area, temp, 'go--')
plt.xticks(rotation=90)
plt.show()
```
