# 판다스(pandas)
+ 파이썬에서 사용하는 데이터분석 라이브러리

+ 행과 열로 이루어진 데이터 객체를 만들어 다룰 수 있게 되며 보다 안정적으로 대용량의 데이터들을 처리하는데 매우 편리한 도구

+ NumPy를 기반으로 구축되었으며 과학 컴퓨팅 환경 내에서 다른 많은 타사 라이브러리와 잘 통합되도록 설계

````
import pandas as pd
````

````
with open('/content/drive/My Drive/PythonWork/PythonBasic/04/부산시기온.csv','r',encoding='utf-8') as fp:
    data = fp.readlines() 

dt = {'일시':[], '평균':[], '최고':[], '최저':[]}
for line in data:
    if line[:2] == '일시' : continue

    line = line.replace('\n','')
    
    item = line.split(',')
    dt['일시'].append(item[0])
    dt['평균'].append(float(item[1]))
    dt['최고'].append(float(item[2]))
    dt['최저'].append(float(item[3]))

print(dt)
print(type(dt))
````

##Series
+ index로 value를 구분하는  1차원 자료구조

````
tempavg = list(dt['평균'])
print(tempavg)
print(type(tempavg))
````

````
#판다스 series 만들기
ds = pd.Series(tempavg)
print(ds)
print(type(ds))
````

````
#인덱스와 값 추출
print(ds.index)
print(ds.keys())
print(ds.values)
````

````
#series 데이터 가져오기
ds1 = ds[11:16]
print(ds1)
````

````
dsDay = dt['일시']
dsDay[:5]
````

````
#series 인덱스 지정하기 
ds = pd.Series(tempavg, index = dsDay)
print(ds)
print(type(ds))
````

````
#인덱스와 값 추출
print(ds.index)
print(ds.keys())
print(ds.values)
````

````
#series 데이터 가져오기 
ds1 = ds['2018-01-01':'2018-12-01']
print(ds1)
````

````
#series 데이터 가져오기 
ds1 = ds[['2018-01-01','2018-12-01']]
print(ds1)
````

````
ds2 = ds[ds > 20]
ds2
````

### series 기초 통계
+ mean() : 평균
+ max() : 최대
+ min() : 최소
+ std() : 표준편차
+ describe() : 요약 통계량 계산

````
print(f'평균 : {ds.mean():0.2f}')
print(f'최대 : {ds.max():0.2f}')
print(f'최소 : {ds.min():0.2f}')
print(f'표준편차 : {ds.std():0.2f}')
print(f'요약 \n{ds.describe() }')
````

````
dsup = ds[ds > ds.mean()]
dsup
````

### series 자주사용되는 메소드
+ replace() : 특정값을 가진 요소 값을 변경
+ equals() : 시리즈와 시리즈가 같은지 비교
+ isin() : 시리즈에 포함된 값이 있는지 확인 
+ drop_duplicates() : 중복값이 없는 시리즈 반환
+ append() : 2개이상의 시리즈 연결 
+ sort_index() : 인덱스로 정렬
 - ascending = False : 역순 정렬 옵션
+ sort_values() : 값으로 정렬
+ to_frame() : 시리즈를 데이터프레임으로 변경

````
ds0 = ds
print(ds0)
````

````
ds0 = ds0.replace(21.3, 21.5)
````

````
print(ds0.equals(ds))
````

````
ds0.isin([21.5]).sum()
````

````
ds0 = ds0.drop_duplicates()
````

````
ds1 = ds[:'2017-06-01']
ds2 = ds['2018-01-01':'2018-06-01']
print(ds1)
print(ds2)
````

````
ds21 = ds2.append(ds1)
ds21
````

````
ds21 = ds21.sort_index()
ds21
````

````
ds21 = ds21.sort_values()
ds21
````

````
ds21 = ds21.to_frame()
print(type(ds21))
````
